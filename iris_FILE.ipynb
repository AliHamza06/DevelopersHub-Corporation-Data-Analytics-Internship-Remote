# Task 1: Exploring and Visualizing the Iris Dataset

**Objective:** Understand how to read, summarize, and visualize a simple dataset using Python libraries like pandas, seaborn, and matplotlib.

**Dataset:** The famous Iris Dataset, which contains measurements of sepal length, sepal width, petal length, petal width, and species for 150 iris flowers.

**Instructions:**
*   Load the dataset using the pandas library.
*   Display dataset structure using `.shape`, `.columns`, and `.head()`.
*   Create basic visualizations:
    *   Scatter plot to analyze relationships between variables.
    *   Histogram to examine data distribution.
    *   Box plot to detect outliers and spread of values.
*   Use `matplotlib` and `seaborn` for visualizations.

### Dataset Loading and Inspection

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
iris_df = sns.load_dataset('iris')

print("Dataset loaded successfully!")
```

```python
# Display dataset structure
print("\nShape of the dataset:", iris_df.shape)
print("\nColumns in the dataset:", iris_df.columns)
print("\nFirst 5 rows of the dataset:\n")
display(iris_df.head())
```

### Basic Visualizations

```python
# Scatter plot to analyze relationships between variables
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris_df, x='sepal_length', y='sepal_width', hue='species', s=100, alpha=0.7)
plt.title('Scatter Plot of Sepal Length vs. Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Species')
plt.show()
```

```python
# Histogram to examine data distribution for a feature (e.g., petal_length)
plt.figure(figsize=(10, 6))
sns.histplot(data=iris_df, x='petal_length', kde=True, hue='species', palette='viridis')
plt.title('Distribution of Petal Length by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# Box plot to detect outliers and spread of values for a feature (e.g., petal_width)
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris_df, x='species', y='petal_width', hue='species', palette='muted', legend=False)
plt.title('Box Plot of Petal Width by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

##  Summary

This notebook demonstrates a basic data exploration and visualization workflow using the Iris dataset. We have covered:

1.  **Data Loading**: Used `seaborn.load_dataset()` to load the Iris dataset into a pandas DataFrame.
2.  **Data Inspection**: Examined the dataset's `shape`, `columns`, and `head()` to understand its structure and content.
3.  **Visualizations**: Created three common types of plots to gain insights:
    *   **Scatter Plot**: To visualize the relationship between `sepal_length` and `sepal_width`, color-coded by `species`.
    *   **Histogram**: To show the distribution of `petal_length` for each `species`.

These steps are fundamental for understanding any new dataset and forming initial hypotheses for further analysis.
